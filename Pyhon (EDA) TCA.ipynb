```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (replace with your actual dataset)
data = {
    'PhoneService': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'MultipleLines': ['No', 'Yes', 'Yes', 'No', 'No'],
    'InternetService': ['DSL', 'Fiber optic', 'DSL', 'No', 'Fiber optic'],
    'OnlineSecurity': ['No', 'Yes', 'No', 'No', 'Yes'],
    'OnlineBackup': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'DeviceProtection': ['No', 'Yes', 'No', 'Yes', 'No'],
    'TechSupport': ['No', 'No', 'Yes', 'Yes', 'No'],
    'StreamingTV': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'StreamingMovies': ['No', 'Yes', 'Yes', 'No', 'Yes'],
    'Churn': ['Yes', 'No', 'No', 'Yes', 'No']  # Add Churn column
}

# Create a DataFrame
df = pd.DataFrame(data)

# Columns to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 
    'StreamingMovies', 'Churn'  # Add Churn here
]

# Validate columns exist in DataFrame
valid_columns = [col for col in columns_to_plot if col in df.columns]

# Set up the figure for subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(14, 16))
axes = axes.flatten()

# Loop through columns and create a count plot for each
for i, column in enumerate(valid_columns):
    sns.countplot(data=df, x=column, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove unused axes if fewer plots than subplots
for j in range(len(valid_columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

#Customers with Fiber Optic internet, no OnlineSecurity, or no TechSupport are more likely to churn, highlighting these as key areas to improve retention. Services like DeviceProtection and Streaming options show no significant impact on churn behavior. Customers with DSL or no internet churn less, suggesting a link between internet type and retention. Focusing on enhancing support services and offering security/backup solutions could effectively reduce churn rates.
